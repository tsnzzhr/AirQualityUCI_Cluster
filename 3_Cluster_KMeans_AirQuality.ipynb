```python
import pandas as pd
air_quality = pd.read_csv("AirQualityUCI.csv")
air_quality
```

```python
air_quality.shape
air_quality.info()
```

```python
aq = air_quality.drop(['Date','Time'], axis=1)
aq.head()
```

```python
aq.shape
aq.info()
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

X = aq.values[:,1:]
print(X)
X = np.nan_to_num(X)
print(X)
clustered = StandardScaler().fit_transform(X)
clustered
```

```python
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score

k_means = KMeans(init = "k-means++", n_clusters = 3, n_init = 10)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
silhouette_score(X, labels)
```

```python
aq["cluster_aq"] = labels
aq.head(5)
```

```python
aq.groupby('cluster_aq')['cluster_aq'].count()

```

```python
aq.groupby('cluster_aq').mean()
```
